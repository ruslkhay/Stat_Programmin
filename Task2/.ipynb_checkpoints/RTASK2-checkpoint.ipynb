Для проверки построить соответствующий график и применить один из формальных методов (на основе расчета коэфициентов модели или критерия Фишера), текстом обосновать почему решили оставить или удалить эффект взаимодействия.
```r
aovm3 <- aov(lm(MPG_City ~ Origin*Type, data = cars)) # Модель с эффектом взаимодействия "Origin:Type"
#summary(aovm3)
#tukey3 <- TukeyHSD(x=aovm3, conf.level=1-p_default)
#summary(lmm3)

plot_intersetion <- function(df) {
    # Sorting df
    na.omit(df) %>% group_by(Origin, Type) %>% 
    summarise(meanloss=mean(MPG_City), se=sd(MPG_City), .groups="rowwise") %>%
    
    ggplot(aes(x=Origin, y=meanloss, colour=Type))+
    geom_line(aes(group=Type)) +
    geom_pointrange(aes(ymin=meanloss - se, ymax= meanloss+ se))
}

plot_intersetion(cars)

# Сохраняю модель
capture.output(summary(aovm3), file = "anova results.txt", append=TRUE)
```

### 6. С помощью оператора Contrast из ghlt или emmeans или aov проверить гипотезу, что средний расход бензина MPG_city не отличается в группе европейских и азиатских седанов от группы американских траков. Проверить, что полученные результаты совпадают с проверкой «в лоб» - изменить соответствующим образом исходную выборку (оставить только модели из двух групп европейских + азиатских седанов vs американские траки) и применить подходящую процедуру ANOVA или t-test.

$$ \mu_{ij}=\alpha_i + \beta_j, \quad \beta_j \in \{Asia(1)\ , Europe(2)\ , USA(3) \}, \quad
\alpha_i \in \{Sedan(1)\ , Truck(2)\} $$

$$ H_0 :\quad \frac{1}{2}(\mu_{11} + \mu_{12}) = \mu_{23} $$

```r
df6 <- confint(emmeans(aovm2, ~ Origin | Type))
AOV<-aov(lm(emmean~ Origin + Type -1,df6))
AOV$coef
tt<-glht(AOV, linfct=matrix(c(1, 1, -2, 2, 0, 0, -2, 0), 1))
summary(tt)

AEs <- subset(cars, Origin %in% c("Asia", "Europe") & Type == 'Sedan', select=c("Origin", "Type", "MPG_City")) # группе европейских и азиатских седанов
Ut <- subset(cars, Origin == "USA" & Type == 'Truck', select=c("Origin", "Type", "MPG_City")) # группе американских траков
t.test(Ut$MPG_City, mu = mean(AEs$MPG_City), conf.level = 1 - p_default)

# Сохраняю модель
capture.output(summary(AOV), file = "anova results.txt", append=TRUE)
```

В обоих случаях р-уровень значимости оказался ниже 0.01, что позволяет отклонить проверяемую гипотизу $H_0$.

```r
transform_data <- function(input_df) {
  # Извлечение уникальных слов из названий строк и создание пустого data.frame
  unique_words <- unique(unlist(strsplit(row.names(input_df), "-")))
  output_df <- data.frame(row.names = unique_words)
  # Создание новых столбцов в выходном data.frame на основе столбцов входного data.frame
  for (col_name in colnames(input_df)) {
    new_col_names <- paste(col_name, unique_words, sep = "_")
    temp_df <- data.frame()
    # Заполнение ячеек выходного data.frame
    for (i in seq_along(unique_words)) {
        # Упорядычиваю названия старых строк, чтобы пары шли в алфавитном порядке 
        sort_pair<- function(name, sep="-"){
            two_names <- unlist(lapply(strsplit(name, sep), FUN=sort)) 
            return(paste(two_names[1], two_names[2], sep=sep))
        }
        # Создаю парные названия с новым словом
        name1 <- unlist(lapply(paste(unique_words,unique_words[i] , sep="-"), FUN=sort_pair))
        name2 <- sort_pair(row.names(input_df)[i])
        cond1 <- grepl(unique_words[i], new_col_names, fixed = TRUE)
        cond2 <- name2!=name1
        temp_df[i, new_col_names]<-ifelse(!cond1&!cond2, input_df[i, col_name], 0)
    }
    output_df <- cbind(output_df, t(temp_df)+temp_df)
    #  output_df <- cbind(output_df, temp_df)
  }
 #output_df <- ifelse(output_df==0, NA, output_df)
  return(output_df)
}

shift_and_concatenate <- function(input_vector) {
  # Создание пустого вектора для хранения результата
  result <- c()
  
  # Добавление исходного вектора в результат
  result <- c(result, input_vector)
  
  # Получение длины исходного вектора
  n <- length(input_vector)
  
  # Инициализация флага для проверки совпадения текущего вектора с исходным
  matching <- FALSE
  
  # Повторение алгоритма, пока полученный вектор не совпадет с исходным
  while (!matching) {
    
      
    # Сдвиг элементов влево
    input_vector <- c(input_vector[-1], input_vector[1])
    
    # Проверка на совпадение текущего вектора с исходным
    matching <- all(input_vector == result[1:n])
    if(matching) break
      
    # Добавление сдвинутого вектора в результат
    result <- c(result, input_vector)
  }
  return(result)
}

restructure_df <- function(input_df) {
  # Извлечение названий столбцов
  col_names <- colnames(input_df)
  
  # Создание списка для хранения объединенных столбцов
  combined_cols <- list()
  
  # Перебор названий столбцов
  for (col_name in col_names) {
    # Извлечение первой буквы названия столбца
    first_letter <- substr(col_name, 1, 1)
    
    # Проверка, существует ли уже объединенный столбец с такой первой буквой
    if (first_letter %in% names(combined_cols)) {
      # Добавление значений текущего столбца к существующему объединенному столбцу
      combined_cols[[first_letter]] <- c(combined_cols[[first_letter]], input_df[[col_name]])
    } else {
      # Создание нового объединенного столбца с текущим столбцом
      combined_cols[[first_letter]] <- input_df[[col_name]]
    }
  }
  
  # Нахождение максимальной длины объединенных столбцов
  max_length <- max(sapply(combined_cols, length))
  
  # Добавление нулевых значений к объединенным столбцам для выравнивания длин
  combined_cols <- lapply(combined_cols, function(x) {
    if (length(x) < max_length) {
      c(x, rep(0, max_length - length(x)))
    } else {
      x
    }
  })
  
  # Создание нового data.frame на основе объединенных столбцов
  output_df <- as.data.frame(combined_cols)
  
  return(output_df)
}
```

```r
shift_and_concatenate_df <- function(input_df) {
  # Создание пустого data.frame для хранения результата
  result_df <- data.frame()
  
  # Добавление исходного data.frame в результат
  result_df <- rbind(result_df, input_df)
  
  # Получение количества строк в исходном data.frame
  n <- nrow(input_df)
  
  # Инициализация флага для проверки совпадения текущего data.frame с исходным
  matching <- FALSE
  
  # Повторение алгоритма, пока полученный data.frame не совпадет с исходным
  while (!matching) {
    # Сдвиг строк вверх
    input_df <- rbind(input_df[-1, ], input_df[1, ])
    
    # Проверка на совпадение текущего data.frame с исходным
    matching <- all(input_df == result_df[1:n, ])
    if (matching) break
    
    # Добавление сдвинутого data.frame в результат
    result_df <- rbind(result_df, input_df)
  }
  
  return(result_df)
}
```
