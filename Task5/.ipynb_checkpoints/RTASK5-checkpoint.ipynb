```r
# Установка рабочего каталога, если файл находится в другой директории
setwd("~/Documents/VMK/MaSP/Data")

# Подключение бибилотек
if (!requireNamespace("vcdExtra", quietly = TRUE)) {
  install.packages("vcdExtra")
}
library(vcdExtra) # библиотека для ф-ии crossval

if (!requireNamespace("crossval", quietly = TRUE)) {
  install.packages("crossval")
}
library(crossval) # библиотека для ф-ии CMHtest

if (!requireNamespace("pROC", quietly = TRUE)) {
  install.packages("pROC")
}
library(pROC) # библиотека для ф-ии ROC


# Загрузка данных из файла
data <- na.omit(read.csv("CARS.csv", header=TRUE, sep=","))

data$Invoice <- as.numeric(gsub("\\$", "", gsub(",", ".", data$Invoice)))
```

## Пункт 1

```r
# Создаем новый столбец Expensive на основе условия
data$Expensive <- ifelse(as.numeric(gsub("[^0-9]", "", data$MSRP)) > 35000, 1, 0)

# Дискретизируем предикторы
data$MPG_Highway_Group <- cut(data$MPG_Highway, breaks = c(0, 24, max(data$MPG_Highway)),
                      labels = c(0, 1))

data$Cylinders_Group <- cut(data$Cylinders, breaks = c(0, 6, max(data$Cylinders)),
                      labels = c(0, 1))

# Проводим тесты
chisq.test(table(data$MPG_Highway_Group, data$Expensive), correct=FALSE)
CMHtest(table(subset(data, select = c('Expensive', 'MPG_Highway_Group', 'Cylinders_Group'))) )
```

## Пункт 2

```r
# Добавляем столбец Cheap
data$Cheap <- ifelse(as.numeric(gsub("[^0-9]", "", data$MSRP)) < 20000, 1, 0)


# Создаем список предикторов
predictors <- c('EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway',
         'Weight','Wheelbase','Length')

# Строим начальную модель логистической регрессии
lm_step <- glm(Cheap ~ 1,
               data = data[,append('Cheap', predictors)],
               family=binomial)

# Определяем функцию для вычисления ROC AUC
ROC <- function(train_x, train_y, test_x, test_y)
{
    model <- glm(train_y ~ . , data=train_x, family=binomial)
    pred_y <- predict(model, test_x)
    return(roc(test_y ~ pred_y, levels = c(0, 1), direction = '<')$auc)
}


# Создаем списки для хранения информации о моделях
predictors_model <- list()
CV_ROC_AUC <- c()
complexity <- 0

# Перебираем предикторы для построения моделей
for (i in 1:length(predictors)) {

    lm_step <- step(lm_step, direction = 'forward',
                     scope = ~  EngineSize + Cylinders + Horsepower + 
                                MPG_City + MPG_Highway + Weight + 
                                Wheelbase + Length,
                     steps = 1,
                     trace = 0 )

    # Имена коэффициентов без Intercept соответствуют именам факторов
    predictor <- list(names(lm_step$coef)[2:length(lm_step$coef)])
    predictors_model <- append(predictors_model, predictor)
    complexity <- complexity + 1
    
    x <- subset(data, select = c(predictors_model[[complexity]]))
    CV_ROC_AUC <- c(CV_ROC_AUC,
                  crossval(ROC,
                           x,
                           data$Cheap,
                           K=5,
                           B=10,
                           verbose=FALSE)$stat)
    
    if (length(predictors_model[[i]]) < i) {
        complexity <- length(predictors_model[[i]])
    }
}

# Список переменных лучшей модели
predictors_model <- predictors_model[[which.max(complexity)]]

# Строим график
plot(1:complexity, CV_ROC_AUC[1:complexity], 
     xlab = "Number of predictors", 
     ylab = "ROC AUC",
     main = "ROC AUC vs Complexity")
```
