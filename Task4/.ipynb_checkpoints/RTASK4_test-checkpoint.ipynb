```r
calculate_MAPE <- function(model) {
  original <- model$fitted.values
  predicted <- model$model[, "MPG_Highway"]
  n <- length(original) 
  mape <- sum(abs(original - predicted) / abs(original)) / n
  if (mape > 0.065) {
    warning("MAPE значение меньше 0.065. Обратите внимание на качество модели.")
  }
  return(mape)
}
```

```r
library(olsrr)
library(caret)
#library(plotly)

# отобраны в model_dev
selected.factors <- c('MPG_Highway', 'log_Weight', 'log_Horsepower', 'log_Length', 'log_Invoice', 'sqrt_Horsepower', 'Horsepower', 'sq_Horsepower', 'log_EngineSize', 'cb_Length', 'sqrt_Length', 'sqrt_Invoice', 'Length', 'sq_Length', 'cb_Weight', 'log_Wheelbase', 'sqrt_Wheelbase', 'cb_Wheelbase', 'cb_Horsepower', 'Invoice', 'sq_Invoice', 'cb_Invoice', 'sqrt_EngineSize', 'EngineSize', 'log_Cylinders', 'sq_EngineSize', 'cb_EngineSize', 'sqrt_Cylinders', 'Cylinders', 'sq_Cylinders', 'cb_Cylinders', 'sq_Weight', 'Weight')
# без отклика
only.factors <- c('log_Weight', 'log_Horsepower', 'log_Length', 'log_Invoice', 'sqrt_Horsepower', 'Horsepower', 'sq_Horsepower', 'log_EngineSize', 'cb_Length', 'sqrt_Length', 'sqrt_Invoice', 'Length', 'sq_Length', 'cb_Weight', 'log_Wheelbase', 'sqrt_Wheelbase', 'cb_Wheelbase', 'cb_Horsepower', 'Invoice', 'sq_Invoice', 'cb_Invoice', 'sqrt_EngineSize', 'EngineSize', 'log_Cylinders', 'sq_EngineSize', 'cb_EngineSize', 'sqrt_Cylinders', 'Cylinders', 'sq_Cylinders', 'cb_Cylinders', 'sq_Weight', 'Weight')

preprocess <- function (cars.sample) {


	cars.sample <- cars.sample[,c('MPG_Highway',
				 'Length',
				 'Weight',
				 'Wheelbase',
				 'Horsepower',
				 'Invoice',
				 'EngineSize',
				 'Cylinders')]

	for (i in colnames(cars.sample)) {
		cars.sample[is.na(cars.sample[i]),i] <- 0.1
	}
	
	cars.sample$Invoice <- as.numeric(gsub('\\$|,', '',cars.sample$Invoice))

	for (i in colnames(cars.sample)) {
		if (i == 'MPG_Highway') {
			next
		}
	
		# производные факторы
		cars.sample[paste('log',i, sep='_')] <- log(cars.sample[i])
		cars.sample[paste('sq',i, sep='_')] <- (cars.sample[i])^2
		cars.sample[paste('cb',i, sep='_')] <- (cars.sample[i])^3
		cars.sample[paste('sqrt',i, sep='_')] <- sqrt(cars.sample[i])

		# нормализация факторов
		cars.sample[i] <- (cars.sample[i] - mean(cars.sample[,i])) / sd(cars.sample[,i])
		cars.sample[paste('log',i, sep='_')] <- (cars.sample[paste('log',i, sep='_')] - mean(cars.sample[,paste('log',i, sep='_')])) / sd(cars.sample[,paste('log',i, sep='_')])
		cars.sample[paste('sq',i, sep='_')] <- (cars.sample[paste('sq',i, sep='_')] - mean(cars.sample[,paste('sq',i, sep='_')])) / sd(cars.sample[,paste('sq',i, sep='_')])
		cars.sample[paste('cb',i, sep='_')] <- (cars.sample[paste('cb',i, sep='_')] - mean(cars.sample[,paste('cb',i, sep='_')])) / sd(cars.sample[,paste('cb',i, sep='_')])
		cars.sample[paste('sqrt',i, sep='_')] <- (cars.sample[paste('sqrt',i, sep='_')] - mean(cars.sample[,paste('sqrt',i, sep='_')])) / sd(cars.sample[,paste('sqrt',i, sep='_')])
	}

	return (cars.sample)
}


fit <- function (cars.table) {
	
	cars.sample <- preprocess(cars.table)
	cars.sample <- subset(cars.sample, select = selected.factors)

	model <- lm(MPG_Highway ~ ., data = cars.sample)
	class(model) <- c('my.lin.reg', 'lm')

	#plot.my.lin.reg <- function ()

	var.imp <- varImp(model)
	var.imp <- var.imp[order(var.imp$Overall),,drop=FALSE]
	most.imp <- rownames(var.imp)
	model$top1.name <- most.imp[length(most.imp)]
	model$top2.name <- most.imp[length(most.imp)-1]

	model$top1.val <- cars.sample[,model$top1.name]
	model$top2.val <- cars.sample[,model$top2.name]
	model$resp <- cars.sample[,'MPG_Highway']
	
	return (model)
}


predict.my.lin.reg <- function (model, test.data) {
	test.data <- preprocess(test.data)
	test.data <- subset(test.data, select = only.factors)
	return (predict.lm(model, test.data))
}

plot.my.lin.reg <- function (model) {
	plot_ly(x=model$top1.val,
	y=model$top2.val,
	z=model$resp,
	type = 'scatter3d',
	mode = "markers",
	marker = list(size = 5, color="blue"),
	width = 900, height = 900)
}


cars.table <- read.csv('CARS.csv')
model <- fit(cars.table)

y.pred <- predict(model, cars.table)
y.true <- cars.table$MPG_Highway

mean(abs((y.true - y.pred)/y.true))

#plot(model)
```
